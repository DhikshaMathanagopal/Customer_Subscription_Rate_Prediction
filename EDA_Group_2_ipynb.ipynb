# Dataset Value Counts

```python
import pandas as pd
import matplotlib.pyplot as plt


labels = data['Subscription Status'].value_counts().index
sizes = data['Subscription Status'].value_counts().values
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85)


centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


ax1.axis('equal')
plt.tight_layout()


plt.title('Subscription Status Distribution', color='#333333', fontsize=14)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

filtered_data = data[data['Gender'].isin(['Male', 'Female'])]
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Gender', data=filtered_data, order=['Male', 'Female'])
colors = ['#1f77b4', '#ff7f0e']
for patch, color in zip(ax.patches, colors):
    patch.set_facecolor(color)
ax.set_title('Count of Each Gender (Male and Female)')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

table = pd.crosstab(index=[data['Gender'], data['Discount Applied']],
                    columns=data['Subscription Status'])
fig, ax = plt.subplots(figsize=(10, 6))
table.plot(kind='bar', stacked=True, ax=ax, color={'Yes': '#4daf4a', 'No': '#e41a1c'})
ax.set_title('Subscription Status by Gender and Discount Application')
ax.set_xlabel('Gender, Discount Applied')
ax.set_ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Subscription Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(data, col="Discount Applied", row="Gender", hue="Subscription Status", height=4, aspect=1.5, margin_titles=True)
g.map_dataframe(sns.histplot, "Review Rating")
g.add_legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data['Review Rating'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Review Ratings')
plt.xlabel('Review Rating')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Subscription Status', y='Review Rating', data=data, palette='coolwarm')
plt.title('Review Ratings by Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('Review Rating')
plt.show()

subscribed = data[data['Subscription Status'] == 'Yes']['Review Rating']
not_subscribed = data[data['Subscription Status'] == 'No']['Review Rating']

t_stat, p_value = ttest_ind(subscribed, not_subscribed, equal_var=False)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

plt.figure(figsize=(12, 6))
sns.countplot(x='Frequency of Purchases', hue='Subscription Status', data=data, palette='Set2', order = data['Frequency of Purchases'].value_counts().index)
plt.title('Frequency of Purchases by Subscription Status')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Subscription Status')
plt.show()

contingency_table = pd.crosstab(data['Frequency of Purchases'], data['Subscription Status'])

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("P-value:", p_value)
```
